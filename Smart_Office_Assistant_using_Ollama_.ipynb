### Task 1: Create Calculator Tool

````python
import ast
import operator

# Supported operators
OPERATORS = {
    ast.Add: operator.add,
    ast.Sub: operator.sub,
    ast.Mult: operator.mul,
    ast.Div: operator.truediv,
    ast.Pow: operator.pow,
    ast.USub: operator.neg,
}

def evaluate(expr):
    """
    Safely evaluate mathematical expressions.
    Example:
        245 * 789
        125 + 560
        100 / 5
    """

    def _eval(node):
        if isinstance(node, ast.Num):
            return node.n

        elif isinstance(node, ast.BinOp):
            return OPERATORS[type(node.op)](
                _eval(node.left),
                _eval(node.right)
            )

        elif isinstance(node, ast.UnaryOp):
            return OPERATORS[type(node.op)](
                _eval(node.operand)
            )

        else:
            raise TypeError("Unsupported expression")

    tree = ast.parse(expr, mode='eval')
    return _eval(tree.body)
````

#### Demonstration of the Calculator Tool

````python
def calculator_tool(query):
    try:
        return str(evaluate(query))
    except Exception as e:
        return f"Calculation Error: {e}"
````

### Task 2: Create Weather Tool

````python
class WeatherTool:
    def __init__(self):
        # Mock weather data
        self.mock_weather_data = {
            "hyderabad": {"temperature": "34°C", "condition": "Sunny"},
            "london": {"temperature": "15°C", "condition": "Cloudy"},
            "new york": {"temperature": "22°C", "condition": "Partly Cloudy"},
            "tokyo": {"temperature": "25°C", "condition": "Rainy"},
        }

    def run(self, location: str) -> str:
        """Fetches mock weather information for a given location.

        Args:
            location: The city for which to get weather information (e.g., 'Hyderabad').

        Returns:
            A string with the mock weather information (e.g., '34°C Sunny') or an error message.
        """
        location_lower = location.lower()
        if location_lower in self.mock_weather_data:
            data = self.mock_weather_data[location_lower]
            return f"{data['temperature']} {data['condition']}"
        else:
            return f"Weather data not available for {location}."

# Instantiate the tool
weather = WeatherTool()
````

#### Demonstration of the Weather Tool

````python
# Test cases
print(f"Weather in Hyderabad: {weather.run('Hyderabad')}")
print(f"Weather in London: {weather.run('London')}")
print(f"Weather in Paris: {weather.run('Paris')}") # Should return data not available
````

### Task 4: Create Unit Conversion Tool

````python
class UnitConversionTool:
    def __init__(self):
        self.conversion_factors = {
            'km_to_m': 1000,
            'm_to_km': 0.001,
            'kg_to_g': 1000,
            'g_to_kg': 0.001,
        }
        self.unit_aliases = {
            'kilometers': 'km', 'kilometer': 'km',
            'meters': 'm', 'meter': 'm',
            'kilograms': 'kg', 'kilogram': 'kg',
            'grams': 'g', 'gram': 'g',
        }

    def run(self, query: str) -> str:
        """Converts units based on the input query.

        Args:
            query: A string representing the conversion request (e.g., '10 km to meters').

        Returns:
            The converted value as a string or an error message.
        """
        try:
            parts = query.lower().split()
            if len(parts) != 4 or parts[2] != 'to':
                return "Invalid query format. Expected: '<value> <from_unit> to <to_unit>'."

            value = float(parts[0])
            from_unit_raw = parts[1]
            to_unit_raw = parts[3]

            # Normalize unit names
            from_unit = self.unit_aliases.get(from_unit_raw, from_unit_raw)
            to_unit = self.unit_aliases.get(to_unit_raw, to_unit_raw)

            conversion_key = f"{from_unit}_to_{to_unit}"
            if conversion_key in self.conversion_factors:
                converted_value = value * self.conversion_factors[conversion_key]
                # Return with the normalized 'to_unit'
                return f"{converted_value} {to_unit}"
            else:
                return f"Unsupported conversion: {from_unit_raw} to {to_unit_raw}."

        except ValueError:
            return "Invalid value. Please provide a numeric value."
        except Exception as e:
            return f"An error occurred: {e}"

# Instantiate the tool
unit_converter = UnitConversionTool()
````

#### Demonstration of the Unit Conversion Tool

````python
# Test cases
print(f"10 km to meters: {unit_converter.run('10 km to meters')}")
print(f"5000 meters to km: {unit_converter.run('5000 meters to km')}")
print(f"2 kg to grams: {unit_converter.run('2 kg to grams')}")
print(f"500 grams to kg: {unit_converter.run('500 grams to kg')}")
print(f"10 miles to km: {unit_converter.run('10 miles to km')}") # Unsupported conversion
print(f"abc km to meters: {unit_converter.run('abc km to meters')}") # Invalid value
````

However, we can demonstrate how to structure the code to make requests to an Ollama server running on a **publicly accessible IP address or a forwarded port**. For the purpose of this demonstration, we'll assume the Ollama server is accessible at a given `OLLAMA_API_BASE_URL`. If you are running Ollama locally, you will need to replace this with your public URL.

````python
import requests

# IMPORTANT: Replace with your actual Ollama API base URL if running locally and exposing it, or a mock URL for demonstration.
# For local testing, you might use ngrok or localtunnel to expose your localhost:11434
OLLAMA_API_BASE_URL = "http://localhost:11434" # Placeholder - **MUST BE REPLACED FOR EXTERNAL ACCESS**

def call_ollama_llm(prompt: str, model: str = "llama2") -> str:
    """Calls a locally hosted Ollama LLM with the given prompt."""
    try:
        response = requests.post(
            f"{OLLAMA_API_BASE_URL}/api/generate",
            json={
                "model": model,
                "prompt": prompt,
                "stream": False
            }
        )
        response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)
        return response.json()["response"]
    except requests.exceptions.ConnectionError:
        return f"Error: Could not connect to Ollama server at {OLLAMA_API_BASE_URL}. Please ensure Ollama is running and accessible from this environment."
    except requests.exceptions.RequestException as e:
        return f"Error calling Ollama LLM: {e}"

# Demonstration of calling the Ollama LLM
print("Attempting to call Ollama LLM...")
llm_response = call_ollama_llm("What is the capital of France?")
print(f"LLM Response: {llm_response}")
````

### Task 6: Create BMI Calculator Tool

This tool will calculate the Body Mass Index (BMI) given a person's weight in kilograms and height in meters.

````python
class BMICalculatorTool:
    def __init__(self):
        pass

    def run(self, weight_kg: float, height_m: float) -> str:
        """Calculates Body Mass Index (BMI).

        Args:
            weight_kg: The person's weight in kilograms.
            height_m: The person's height in meters.

        Returns:
            A string indicating the calculated BMI and its category, or an error message.
        """
        try:
            if not isinstance(weight_kg, (int, float)) or not isinstance(height_m, (int, float)):
                return "Error: Weight and height must be numeric values."
            if weight_kg <= 0 or height_m <= 0:
                return "Error: Weight and height must be positive values."

            bmi = weight_kg / (height_m ** 2)
            category = ""
            if bmi < 18.5:
                category = "Underweight"
            elif 18.5 <= bmi < 24.9:
                category = "Normal weight"
            elif 25 <= bmi < 29.9:
                category = "Overweight"
            else:
                category = "Obesity"
            return f"Your BMI is {bmi:.2f}, which falls into the '{category}' category."
        except Exception as e:
            return f"An error occurred during BMI calculation: {e}"

# Instantiate the tool
bmi_calculator = BMICalculatorTool()
````

#### Demonstration of the BMI Calculator Tool

````python
# Test cases
print(f"BMI for 70 kg and 1.75 m: {bmi_calculator.run(70, 1.75)}")
print(f"BMI for 85 kg and 1.80 m: {bmi_calculator.run(85, 1.80)}")
print(f"BMI for 50 kg and 1.60 m: {bmi_calculator.run(50, 1.60)}")
print(f"BMI for 0 kg and 1.70 m: {bmi_calculator.run(0, 1.70)}") # Error case
print(f"BMI for 70 kg and 'invalid' m: {bmi_calculator.run(70, 'invalid')}") # Error case
````
