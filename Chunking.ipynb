```python
text = """
Employees receive 12 casual leaves annually.
Employees can work from home twice per week.
Travel expenses are reimbursed within 30 days.
Medical insurance is provided to all employees.
"""
```

```python
chunk_size = 50
chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
for chunk in chunks:
  print(chunk)
  print('-'*50)
```

# Recursive Chunking

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size=50, chunk_overlap=10)
chunks = splitter.split_text(text)
for i,chunk in enumerate(chunks):
  print(f'Chunk (i+1)')
  print(chunk)
  print('-'*50)
```

# Basic Sentence Chunking using NLTK

```python
!pip install nltk
import nltk
nltk.download('all')
```

```python
from nltk.tokenize import sent_tokenize
chunks = sent_tokenize(text)
for i,chunk in enumerate(chunks):
  print(f'Chunk (i+1)')
  print(chunk)
  print('-'*50)
```

```python
sentences = sent_tokenize(text)
chunk_size = 2
chunks = []

for i in range(0,len(sentences),chunk_size):
  chunk = ' '.join(sentences[i:i+chunk_size])
  chunks.append(chunk)

for id,chunk in enumerate(chunks):
  print(f'Chunk {id}:')
  print(chunk)
  print('-'*50)
```

# Semantic Chjunking Using Sentence Transformers

```python
!pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
senteces = sent_tokenize(text)
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
threshold = 0.5
chunks = []
current_chunk = [sentences[0]]
for i in range(1,len(sentences)):
  similarity = cosine_similarity([embeddings[0]],embeddings[1:])[0][0]

  if similarity > threshold:
    current_chunk.append(sentences[i])
  else:
    chunks.append(' '.join(current_chunk))
    current_chunk = [sentences[i]]

chunks.append(' '.join(current_chunk))

for id,chunk in enumerate(chunks):
  print(f'\nChunk {id}:')
  print(chunk)
  print('-'*50)
```
