#Build a Manual AI Agent in Python


## Step 1 – Calculator Tool

```python
def calculator_tool(expression):
    try:
        return eval(expression)
    except Exception as e:
        return str(e)
```

## Step 2 – Weather Tool

```python
def weather_tool(city):
    weather_data = {
        'hyderabad':'34°C, Sunny',
        'vijayawada':'36°C, Cloudy',
        'delhi':'40°C, Hot'
    }
    return weather_data.get(city.lower(),'Weather Data Not Available')
```

## Step 3 – Time Tool

```python
from datetime import datetime

def time_tool():
    return datetime.now().strftime('%d-%m-%Y %H:%M:%S')
```

## Step 4 – Stocks Tool

```python
def stock_tool(symbol):
    stock_data = {
        'msft': 'MSFT: $420.69',
        'goog': 'GOOG: $175.30',
        'appl': 'AAPL: $210.00'
    }
    return stock_data.get(symbol.lower(), 'Stock Data Not Available')
```

## Step 5 – Health Tool

```python
def health_tool(query):
    health_info = {
        'fever': 'Common symptoms include high temperature, chills, and body aches.',
        'headache': 'Often relieved by rest, hydration, and over-the-counter pain relievers.',
        'flu': 'Caused by influenza viruses, symptoms include fever, cough, sore throat, and muscle aches.'
    }
    for keyword, info in health_info.items():
        if keyword in query.lower():
            return info
    return 'Health information not available for this query.'
```

## Step 6 – Military Tool

```python
def military_tool(query):
    military_facts = {
        'army': 'The United States Army is the land warfare service branch of the U.S. Armed Forces.',
        'navy': 'The United States Navy is the maritime warfare service branch of the U.S. Armed Forces.',
        'air force': 'The United States Air Force is the aerial warfare service branch of the U.S. Armed Forces.'
    }
    for keyword, fact in military_facts.items():
        if keyword in query.lower():
            return fact
    return 'Military information not available for this query.'
```

## Step 7 – Science Tool

```python
def science_tool(topic):
    science_facts = {
        'gravity': 'Gravity is a fundamental force of nature that attracts any objects with mass.',
        'photosynthesis': 'Photosynthesis is the process used by plants to convert light energy into chemical energy.',
        'relativity': 'Einstein\'s theory of relativity deals with the relationship between space and time.'
    }
    for keyword, fact in science_facts.items():
        if keyword in topic.lower():
            return fact
    return 'Science information not available for this topic.'
```

## Step 8 – Movies Tool

```python
def movie_tool(title):
    movie_details = {
        'inception': 'Inception (2010) - Directed by Christopher Nolan, starring Leonardo DiCaprio.',
        'interstellar': 'Interstellar (2014) - Directed by Christopher Nolan, starring Matthew McConaughey.',
        'the matrix': 'The Matrix (1999) - Directed by The Wachowskis, starring Keanu Reeves.'
    }
    for keyword, detail in movie_details.items():
        if keyword in title.lower():
            return detail
    return 'Movie details not available for this title.'
```

```python
def agent(user_query):

    print('\nTHOUGHT:')

    if any(op in user_query for op in ['+','-','*','/']):

        print('This looks like a calculation.')
        print('\nACTION: Using Calculator Tool')

        result = calculator_tool(user_query)

        print('\nOBSERVATION:')
        print(result)

        return result

    elif 'weather' in user_query.lower():

        print('Need weather information.')

        city = user_query.lower().split()[-1]

        print('\nACTION: Using Weather Tool')

        result = weather_tool(city)

        print('\nOBSERVATION:')
        print(result)

        return result

    elif 'time' in user_query.lower():

        print('Need current time.')

        print('\nACTION: Using Time Tool')

        result = time_tool()

        print('\nOBSERVATION:')
        print(result)

        return result

    elif 'stock' in user_query.lower():

        print('Need stock information.')
        # Assuming the stock symbol is the last word in the query
        symbol = user_query.lower().split()[-1]

        print('\nACTION: Using Stock Tool')

        result = stock_tool(symbol)

        print('\nOBSERVATION:')
        print(result)

        return result

    elif 'health' in user_query.lower() or 'fever' in user_query.lower() or 'headache' in user_query.lower() or 'flu' in user_query.lower():

        print('Need health information.')

        print('\nACTION: Using Health Tool')

        result = health_tool(user_query)

        print('\nOBSERVATION:')
        print(result)

        return result

    elif 'military' in user_query.lower() or 'army' in user_query.lower() or 'navy' in user_query.lower() or 'air force' in user_query.lower():

        print('Need military information.')

        print('\nACTION: Using Military Tool')

        result = military_tool(user_query)

        print('\nOBSERVATION:')
        print(result)

        return result

    elif 'science' in user_query.lower() or 'gravity' in user_query.lower() or 'photosynthesis' in user_query.lower() or 'relativity' in user_query.lower():

        print('Need science information.')

        print('\nACTION: Using Science Tool')

        result = science_tool(user_query)

        print('\nOBSERVATION:')
        print(result)

        return result

    elif 'movie' in user_query.lower() or 'inception' in user_query.lower() or 'interstellar' in user_query.lower() or 'matrix' in user_query.lower():

        print('Need movie information.')
        # Extract movie title, simple approach for demonstration
        if 'about' in user_query.lower():
            title = user_query.lower().split('about')[-1].strip()
        else:
            title = user_query.lower().split('movie')[-1].strip()

        print('\nACTION: Using Movie Tool')

        result = movie_tool(title)

        print('\nOBSERVATION:')
        print(result)

        return result

    return 'I do not know how to handle this request.'
```
